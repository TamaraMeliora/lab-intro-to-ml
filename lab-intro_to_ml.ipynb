```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check for missing values**

```python
print(spaceship.isnull().sum())
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship_clean = spaceship.dropna()
```

**KNN**

K Nearest Neighbors is a distance based algorithm, and requeries all **input data to be numerical.**

Let's only select numerical columns as our features.

```python

numerical_features = spaceship_clean.select_dtypes(include=['int64', 'float64'])


print(numerical_features.columns)
```

And also lets define our target.

```python
# target will be  - Transported
y = spaceship_clean['Transported']

# features
X = spaceship_clean.select_dtypes(include=['int64', 'float64'])

print("Size X:", X.shape)
print("Size y:", y.shape)
```

**Train Test Split**

Now that we have split the data into **features** and **target** variables and imported the **train_test_split** function, split X and y into X_train, X_test, y_train, and y_test. 80% of the data should be in the training set and 20% in the test set.

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,
    random_state=0
)

# Проверим размеры полученных наборов
print("Size X_train:", X_train.shape)
print("Size X_test:", X_test.shape)
print("Size y_train:", y_train.shape)
print("Size y_test:", y_test.shape)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

You need to choose between **Classificator** or **Regressor**. Take into consideration target variable to decide.

Initialize a KNN instance without setting any hyperparameter.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(3)
```

Fit the model to your data.

```python
knn.fit(X_train, y_train) 
```

Evaluate your model.

```python
from sklearn.metrics import accuracy_score

# Запрос значений признаков у пользователя
age = float(input("Enter Age: "))
room_service = float(input("Enter RoomService amount: "))
food_court = float(input("Enter FoodCourt amount: "))
shopping_mall = float(input("Enter ShoppingMall amount: "))
spa = float(input("Enter Spa amount: "))
vr_deck = float(input("Enter VRDeck amount: "))

# Формируем массив признаков для одного нового пассажира
new_passenger = [[age, room_service, food_court, shopping_mall, spa, vr_deck]]
```

```python
# Делаем предсказание для нового пассажира
prediction = knn.predict(new_passenger)
print("Prediction (Transported):", prediction[0])
```

```python
# Предсказания на тестовой выборке
y_pred = knn.predict(X_test)
print(y_pred)
```

```python
# Оцениваем точность модели на тесте
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy:.4f}")
```
